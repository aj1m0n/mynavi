```python
import pandas as pd
import numpy as np
```

```python
trainData = pd.read_csv("./2new_traindata.csv",index_col=0)
testData = pd.read_csv("./2new_testdata.csv",index_col=0)
```

```python
match_reports = pd.read_csv('./match_reports.csv')
```

```python
match_reports.head()
```

```python
trainscore = []
testscore =[]
for i in range(len(match_reports)):
    if i<len(trainData):
        trainscore.append(match_reports['home_team_score'][i] - match_reports['away_team_score'][i])
    else:
        testscore.append(match_reports['home_team_score'][i]-match_reports['away_team_score'][i])
```

```python
trainData['score'] = trainscore
for i in range(18):
    testscore.append(0)
testData['score'] = testscore
```

## score

```python
trainscores = []
testscores = []
for i in range(len(trainData)):
    if trainData['score'][i] > 0:
        trainscores.append(2)
    elif trainData['score'][i] == 0:
        trainscores.append(1)
    else:
        trainscores.append(0)
        
for i in range(len(testData)):
    
    if testData['score'][i] > 0:
        testscores.append(2)
    elif testData['score'][i] == 0:
        testscores.append(1)
    else:
        testscores.append(0)
```

```python
trainData['scores'] = trainscores
testData['scores'] = testscores
```
